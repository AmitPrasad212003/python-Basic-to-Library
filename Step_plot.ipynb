```python
import matplotlib.pyplot as plt

x = ["day1","day2","day3","day4","day5"]
y = [30,35,45,56,67]

plt.step(x,y)
plt.show()
```

```python
plt.step(x,y,where = "post")
```

```python
plt.step(x,y, where = "mid", marker = "o")
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

data = pd.read_csv("bonus.csv")
df = pd.DataFrame(data)
# print(df)

group = df.groupby("Team").agg({"Salary" :"sum"})
print(group)

plt.step(group.index, group["Salary"],where = "post", marker = "o")
plt.show()
```
