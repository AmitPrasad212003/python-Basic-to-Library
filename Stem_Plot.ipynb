```python
import matplotlib.pyplot as plt

a = [20,30,40,50,0,30,40,30,30,70,80,70,80,82,88,40,20]
plt.stem(a,linefmt = "--",markerfmt = "^",bottom = 5)
plt.show()
```

```python
plt.stem(a,linefmt = "--",markerfmt = "*",bottom = 5,orientation = "horizontal")
plt.show()
```
