```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
print(data)

# cat plot stands for catrgorical plot
sns.catplot(data = data,x = "day", y = "tip",hue = "sex")
plt.show()
```

```python
sns.catplot(data = data,x = "day", y = "tip",hue = "sex",kind = "bar")
plt.show()
```

```python
sns.catplot(data = data,x = "day", y = "tip",hue = "sex",kind = "violin")
plt.show()
```

```python
sns.catplot(data = data,x = "day",hue = "sex",kind = "count")
plt.show()
```
