```python
import seaborn as sns
import matplotlib.pyplot as plt

# https://github.com/mwaskom/seaborn-data


data = sns.load_dataset("tips")
print(data)
sns.barplot(data = data, x = "day", y = "tip", hue= "sex",palette = "viridis",order = ["Sun","Sat","Fri","Thur"], errorbar = ("ci",0))
plt.show()
```
