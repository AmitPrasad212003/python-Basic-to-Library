```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
print(x)
print(y)
plt.scatter(x,y)
plt.show()
```

```python
x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)

plt.scatter(x,y,marker = "*",color = 'r')
plt.show()
```

```python
x = np.random.randint(1,10,50)
y = np.random.randint(10,100,50)
colors = np.random.randint(10,100,50)
size = np.random.randint(10,100,50)
plt.scatter(x,y,marker = "^",cmap = "viridis",c = colors ,s = size)
plt.colorbar()
plt.show()
```

```python
plt.scatter(x,y,marker = "^",cmap = "any",c = colors)
plt.colorbar()
plt.show()
```

```python
plt.scatter(x,y,marker = "*",cmap = "hot",c = colors)
plt.colorbar()
plt.show()
```
