```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,50,30,55,80,50,90]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days ,NOP1,NOP2,NOP3)
plt.show()
```

```python
plt.stackplot(days ,NOP1,NOP2,NOP3,baseline = "sym",colors= ["red","yellow","black"])
plt.show()
```

```python
plt.stackplot(days ,NOP1,NOP2,NOP3,labels = ["week1","week2","week3"])
plt.legend()
plt.show()
```
