```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")

sns.pairplot(data,hue ="day")
plt.show()
```

```python
sns.pairplot(data,hue ="day", diag_kind = "bar")
plt.show()
```

```python
sns.pairplot(data,hue ="day", diag_kind = "plot")
plt.show()
```

```python
sns.pairplot(data,hue ="day", diag_kind = "hist")
plt.show()
```
