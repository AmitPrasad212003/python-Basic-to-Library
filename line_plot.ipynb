```python
import matplotlib.pyplot as plt

x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]

plt.plot(x,y)
plt.show()
```

```python
# plt.plot(x,y,marker = "o")
# plt.plot(x,y,marker = "D") 
# plt.plot(x,y,marker = "*")
plt.plot(x,y,marker = "^")
# plt.plot(x,y,marker = "^",ls = ":")
plt.plot(x,y,marker = "^",ls = "--",color = 'm')



plt.show()
```

```python
x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]
y1 = [500,450,300,250,320]

plt.plot(x,y,marker = "^",ls = "--",color = 'r',label = "week 1")
plt.plot(x,y1,marker = "*",ls = "-",color = 'g',label = "week 2",alpha = 0.5)
plt.legend()
plt.show()
```
