```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
print(data)

gp = data.groupby("day").agg({"tip": "mean"})
sns.heatmap(gp )
plt.show()
```

```python
data = pd.read_csv("bonus.csv")
# print(data)
gp = data.groupby("Team").agg({"Salary": "mean"})
sns.heatmap(gp)
plt.show()
```
