```python
#### Bar plot ########

import matplotlib.pyplot as plt
# 13:45:47 time line

y =[98,67,88,95,88]
x = ["part1","part2","part3","part4","part5"]

plt.bar(x,y)
plt.xlabel("Parts of Harry Potter")
plt.ylabel("Populatity")
plt.title("Populaties of Different parts of Harry potter")

plt.show()
```

```python
y =[98,67,88,95,88]
x = ["part1","part2","part3","part4","part5"]

plt.bar(x,y,color="red" )
plt.xlabel("Parts of Harry Potter",fontsize = 15)
plt.ylabel("Populatity",fontsize = 15)
plt.title("Populatities of Different parts of Harry potter",fontsize = 20)
plt.show()
```

```python
colors = ["red","green","yellow","blue","orange"]
plt.bar(x,y,color = colors)
plt.savefig("bar.png")

plt.show()
```

```python
plt.bar(x,y,color = colors,edgecolor = "black")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("bonus.csv")
df=pd.DataFrame(data)
print(df)

plt.bar(df["Gender"],df["Salary"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("bonus.csv")
df=pd.DataFrame(data)
print(df)

grouped = df.groupby("Gender")["Salary"].sum()
print(grouped)

plt.bar(grouped.index,grouped.values)
plt.show()
```
