```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd

data = sns.load_dataset("tips")
# print(data)

sns.relplot(data = data, x = "tip",y = "total_bill", hue = "sex", col = "day", size = "smoker")
plt.show()
```
