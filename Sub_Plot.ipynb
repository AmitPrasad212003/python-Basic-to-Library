```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,34,56,23,45]
plt.figure(figsize = [6,4])
plt.subplot(1,3,1) # rows, columns, chartnum
plt.title("Age")
plt.plot(x,y)
# plt.show()
x = [5,6,7,8,9]
y = [67,50,66,56,82]
plt.subplot(1,3,2)
plt.title("Weight")
plt.plot(x,y)

x = [1,3,5,7,9]
y = [57,20,16,76,2]
plt.subplot(1,3,3)
plt.title("Chest")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,34,56,23,45]
# plt.figure(figsize = [6,4])
plt.subplot(2,2,1)
plt.title("Age")
plt.plot(x,y)
# plt.show()
x = [5,6,7,8,9]
y = [67,50,66,56,82]
plt.subplot(2,2,2)
plt.title("Weight")

plt.bar(x,y)

x = [1,3,5,7,9]
y = [57,20,16,76,2]
plt.subplot(2,2,3)
plt.title("Chest")

plt.step(x,y)

x = [1,3,5,7,9]
y = [57,20,16,76,2]
plt.subplot(2,2,4)
plt.title("NUmbers")

plt.stem(x,y)

plt.suptitle("Sigma_Male")
plt.show()
```
