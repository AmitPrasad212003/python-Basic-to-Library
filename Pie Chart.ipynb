```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
x = [22,35,30,3,10]

c = ["red","green","blue","pink","orange"]
ex = [0,0,0,0,0.2]
plt.pie(x,labels = brands,  colors = c,explode = ex,shadow = True,autopct = "%.3f",startangle =90)
plt.savefig("pie.pdf",facecolor = "yellow",pad_inches = 0.5  , bbox_inches = "tight")
plt.show()
```
